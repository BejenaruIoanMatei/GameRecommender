### TF-IDF
- TF (Term Frequency): cate cuvinte apar in descrierea unui joc
- IDF (Inverse Document Frequency): reduce importanta cuvintelor care apar frecvent in descrierile altor jocuri
- TF x IDF si obtinem cuvinte care au frecventa mare in descriere dar care nu apar des in alte documente

### Cosine Similarity
- Cosine similarity practic masoara unghiul dintre doi vectori
- Daca am un unghi apropiat de 1 inseamna ca doua jocuri sunt foarte similare, deci un cosine mai mare inseamna jocuri mai similare

- Merge bine pe descrieri de jocuri
- Se focuseaza pe directia vectorilor

Deci folosim cosine similarity ca sa masuram asemanarile dintre jocuri

```python
import numpy as np
```

```python
def jaccard_similarity(set1, set2): 
    intersection = len(set1 & set2)
    union = len(set1 | set2)
    return intersection / union if union != 0 else 0

# recomanda jocuri similare cu un titlu dat
#   -> cosine similarity
#   -> jaccard similarity

def recommend_game(title, n_recommendation=5, alpha=0.3):
    title = title.lower().strip()

    if 'name_norm' not in data.columns:
        data['name_norm'] = data['Name'].str.lower().str.strip()

    if title not in data['name_norm'].values:
        return {"error": "Game not found. Please check the name and try again."}

    g_idx = data[data['name_norm'] == title].index[0]

    # Jaccard
    jaccard_scores = [
        0 if idx == g_idx else jaccard_similarity(
            data.loc[g_idx, 'Tags'], data.loc[idx, 'Tags']
        )
        for idx in range(len(data))
    ]
    jaccard_scores = np.array(jaccard_scores)
    if np.max(jaccard_scores) > 0:
        jaccard_scores /= np.max(jaccard_scores)

    # Cosine
    cosine_scores = cosine_similarity(games_vector[g_idx], games_vector).flatten()
    if np.max(cosine_scores) > 0:
        cosine_scores /= np.max(cosine_scores)

    # Scor final
    final_score = alpha * cosine_scores + (1 - alpha) * jaccard_scores
    recomm_idx = final_score.argsort()[::-1][1:n_recommendation + 1]

    # Rezultate structurate
    recommendations = []
    for idx in recomm_idx:
        game_name = data.iloc[idx]['Name']
        game_image = data.iloc[idx].get('Header image', '')  # fallback gol dacă nu există
        recommendations.append({
            "name": game_name,
            "image": game_image
        })

    return {
        "input": title,
        "recommendations": recommendations
    }

```

```python
title = 'battlefield 1'
print(recommend_game(title))
```

```python
# in esenta face acelasi lucru ca recommend_game, doar ca primeste ca input mai multe jocuri simultan
# combina scorurile de similaritate ale fiecaruia cu toate celelalte si recomanda un joc similar cu toate 3

def recommend_game_multiple(titles, alpha=0.3):
    titles = [t.lower().strip() for t in titles]
    if 'name_norm' not in data.columns:
        data['name_norm'] = data['Name'].str.lower().str.strip()

    for title in titles:
        if title not in data['name_norm'].values:
            return f"Game '{title}' not found. Please check the names and try again."

    idxs = [data[data['name_norm'] == t].index[0] for t in titles]

    jaccard_scores_list = []
    for g_idx in idxs:
        jaccard_scores = []
        for idx in range(len(data)):
            if idx == g_idx:
                jaccard_scores.append(0)
            else:
                jaccard_scores.append(jaccard_similarity(data.loc[g_idx, 'Tags'], data.loc[idx, 'Tags']))
        jaccard_scores = np.array(jaccard_scores)
        if np.max(jaccard_scores) > 0:
            jaccard_scores /= np.max(jaccard_scores)
        jaccard_scores_list.append(jaccard_scores)

    cosine_scores_list = []
    for g_idx in idxs:
        cosine_scores = cosine_similarity(games_vector[g_idx], games_vector).flatten()
        if np.max(cosine_scores) > 0:
            cosine_scores /= np.max(cosine_scores)
        cosine_scores_list.append(cosine_scores)

    # media intre scoruri
    jaccard_mean = np.mean(jaccard_scores_list, axis=0)
    cosine_mean = np.mean(cosine_scores_list, axis=0)

    final_score = alpha * cosine_mean + (1 - alpha) * jaccard_mean

    for idx in idxs: # asta ca sa nu primim aceleasi jocuri date ca input :))
        final_score[idx] = -1

    best_idx = np.argmax(final_score)
    recommended_game = data.iloc[best_idx]['Name']

    return f"Recommended game similar to {', '.join(titles)} is: {recommended_game}"
```

```python
titles = ['Forza Horizon 5', 'Scrap', 'need for speed heat']
#titles = ['Karate Cat','Sword of Destiny','Consumed Awakening']
print(recommend_game_multiple(titles))
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn_model = NearestNeighbors(n_neighbors=6, metric='cosine')
nn_model.fit(games_vector)

def recommend_game_nn(title, n_recommendation=5):
    title = title.lower().strip()

    if 'name_norm' not in data.columns:
        data['name_norm'] = data['Name'].str.lower().str.strip()

    if title not in data['name_norm'].values:
        return {"error": "Game not found. Please check the name and try again."}

    g_idx = data[data['name_norm'] == title].index[0]
    game_vector = games_vector[g_idx]

    distances, indices = nn_model.kneighbors(game_vector, n_neighbors=n_recommendation + 1)
    recommended_indices = indices.flatten()[1:]

    recommendations = []
    for idx in recommended_indices:
        game_name = data.iloc[idx]['Name']
        game_image = data.iloc[idx].get('Header image', '')
        recommendations.append({
            "name": game_name,
            "image": game_image
        })

    return {
        "input": title,
        "recommendations": recommendations
    }
```
